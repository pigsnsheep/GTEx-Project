```python
gtx_data = pd.read_csv("./gene_tpm_2017-06-05_v8_colon_sigmoid.gct", sep="\t", header=2)
```

```python
gene_ids = gtx_data['Name']
gene_ids
```

```python
gtx_data # 56200 rows × 376 columns
```

```python
gtx_data.columns
```

```python
# drop columns
gtx_data = gtx_data.drop(columns=['Name', 'Description', 'id'])
gtx_data
```

```python
# transform so genes as columns (variables), samples as rows (treatment?)  
gtx_data_t = gtx_data.T
gtx_data_t
```

```python
# apply PCA
pca = decomposition.PCA(n_components=2)
gtx_data_t_pca = pca.fit_transform(gtx_data_t)
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=gene_ids.T)
loadings
```

```python
loadings_pc1 = loadings.abs().sort_values(by=['PC1'], ascending=False)
loadings_pc1
```

```python
loadings_pc2 = loadings.abs().sort_values(by=['PC2'], ascending=False)
loadings_pc2
```
